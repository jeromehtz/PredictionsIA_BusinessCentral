```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
dfSales = pd.read_csv("items_sales_data.csv", sep=';')
dfSalesGlobalQuantity = pd.read_csv("items_sales_data.csv", sep=';')
dfSales = dfSales.drop(columns=['GlobalQuantity'])
dfSalesGlobalQuantity = dfSalesGlobalQuantity.drop(columns=['UnitPrice', 'LineAmount', 'LastQuantity', 'DocumentDate'])
dfSales
```

```python
dfSalesGlobalQuantity
```

```python
''' dupliquer les lignes de dfSales en respectant les points suivants (1000 lignes au total)
        - garder le unitPrice
        - changer la quantité pour chaque ligne dupliquée
        - recalculer la quantité globale pour chaque article
        - mettre des dates supérieures pur calculer le profit sur une horizon de min. 5 ans
        - recalculer le montant ligne pour chaque article (unitPrice * last quantity)

'''

def dupliquer_lignes(df):
    List = []
    for index, row in df.iterrows():
        for i in range(20):
            row_dict = row.to_dict()
            List.append(row_dict)
    return pd.DataFrame(List)

        
dfSalesCopy = dupliquer_lignes(dfSales)
dfSalesCopy
```

```python
dfSalesCopy["LastQuantity"] = np.random.randint(1, 20, size=len(dfSalesCopy))
dfSalesCopy["LineAmount"] = round(dfSalesCopy["UnitPrice"] * dfSalesCopy["LastQuantity"])
# dfSalesCopy["GlobalQuantity"] = dfSalesCopy["GlobalQuantity"] + dfSalesCopy["LastQuantity"]
dfSalesCopy
```

```python
dfSalesGlobalBenefit = (
    dfSalesCopy.groupby('ItemNo', as_index=False)['LastQuantity']
    .sum()
    .rename(columns={'LastQuantity': 'GlobalQuantity'})
)
dfSalesGlobalBenefit['UnitPrice'] = dfSales['UnitPrice']
dfSalesGlobalBenefit['GlobalLineAmount'] = round(dfSalesGlobalBenefit['GlobalQuantity'] * dfSales['UnitPrice'])
dfSalesGlobalBenefit

```

```python
dfSalesCopy
```

```python
today = datetime.today()
ten_years_later = today + timedelta(days=365*10)

def random_future_date(start, end):
    delta = end - start
    random_days = random.randint(0, delta.days)
    return start + timedelta(days=random_days)

dfSalesCopy['DocumentDate'] = [
    random_future_date(today, ten_years_later) 
    for _ in range(len(dfSalesCopy))
]
dfSalesCopy.rename(columns={'LastQuantity': 'Quantity'}, inplace=True)
dfSalesCopy
```

```python
dfSalesGlobalBenefit['DocumentDate'] = (
    dfSalesCopy.assign(DocumentDate=pd.to_datetime(dfSalesCopy['DocumentDate'], format='%d/%m/%Y'))
    .groupby('ItemNo')['DocumentDate']
    .max()
    .dt.strftime('%d/%m/%Y')
    .values
)
dfSalesCopy['DocumentDate'] = pd.to_datetime(dfSalesCopy['DocumentDate']).dt.strftime('%d/%m/%Y')
```

```python
dfSalesCopy.to_csv("output/sales_data.csv", sep=';', encoding='utf-8-sig')
dfSalesCopy
```
